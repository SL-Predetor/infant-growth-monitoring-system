```python
import tempfile
from matplotlib import pyplot
from matplotlib.patches import Rectangle
from matplotlib.patches import Circle
from mtcnn.mtcnn import MTCNN
import math
from PIL import Image
import os
import numpy as np
```

```python
input_directory = r'C:\Users\Yasindu\Desktop\Stuff\1.SLIIT\Research\Project\test_data\rawimg\Non_Autistic'
dataS = os.listdir(input_directory)

output_directory = r'C:\Users\Yasindu\Desktop\Stuff\1.SLIIT\Research\Project\test_data\prepImg\step1\Non_Autistic'
if not os.path.exists(output_directory):
    os.makedirs(output_directory)
```

```python
detector = MTCNN()
count = 0
margin = 0.1
```

```python
# Function to draw and save cropped faces
def draw_faces(filename, result_list):
    global count
    # Load the image
    img = Image.open(filename)
    Iwidth, Iheight = img.size

    # Loop through all detected faces
    for i in range(len(result_list)):
        # Get the bounding box coordinates
        x1, y1, width, height = result_list[i]['box']
        x2, y2 = x1 + width, y1 + height

        # Apply margin to the bounding box
        x1 = max(math.floor(x1 * (1 - margin)), 0)
        y1 = max(math.floor(y1 * (1 - margin)), 0)
        x2 = min(math.floor(x2 * (1 + margin)), Iwidth)
        y2 = min(math.floor(y2 * (1 + margin)), Iheight)

        # Crop the face from the image
        im_crop = img.crop((x1, y1, x2, y2))

        # Save the cropped image with a unique filename
        im_crop.save(os.path.join(output_directory, f"{count}.jpg"), quality=100)

        # Print the current count to check during each iteration
        print(f"Saved image {count}.jpg")

        # Increment the count
        count += 1
```

```python
for file in dataS:
    file_path = os.path.join(input_directory, file)  # Full file path for each image
    if file.lower().endswith(('.jpg', '.jpeg', '.png')):  # Process only image files
        try:
            print(f"Processing file: {file_path}")  # Print the file being processed

            # Open the image using PIL and convert to RGB to ensure the correct format
            img = Image.open(file_path).convert('RGB')

            # Convert image to a NumPy array
            img_array = np.array(img)

            # Ensure image is in the correct format (uint8) and has shape (height, width, 3)
            img_array = np.clip(img_array, 0, 255).astype(np.uint8)

            # Ensure the image has the right shape for MTCNN (add batch dimension)
            pixels = np.expand_dims(img_array, axis=0)  # Shape: (1, height, width, 3)

            # Detect faces in the image
            faces = detector.detect_faces(pixels[0])  # Pass only the first image in the batch
            
            # Draw and save the faces
            draw_faces(file_path, faces)

            # After processing, delete the image (optional)
            os.remove(file_path)
            print(f"Deleted image: {file_path}")
            
        except Exception as e:
            print(f"Error processing {file}: {e}")
    else:
        print(f"Skipping non-image file: {file}")
```
