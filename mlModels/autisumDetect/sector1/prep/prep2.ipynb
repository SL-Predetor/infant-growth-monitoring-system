```python
import hashlib
import os
import shutil
```

```python
def file_hash(filepath):
    """Generate MD5 hash for the file"""
    with open(filepath, 'rb') as f:
        return hashlib.md5(f.read()).hexdigest()
```

```python
os.chdir(r'C:\Users\Yasindu\Desktop\Stuff\1.SLIIT\Research\Project\data\imgData\prepImg\step3(S1+S2.2)\Non_Autistic')
file_list = os.listdir()
destination_folder = r'C:\Users\Yasindu\Desktop\Stuff\1.SLIIT\Research\Project\data\imgData\prepImg\step4\Non_Autistic'
if not os.path.exists(destination_folder):
    os.makedirs(destination_folder)
```

```python
# Initialize the dictionary and list for duplicates
duplicates = []
hash_keys = dict()
```

```python
# Loop through each file in the directory
for index, filename in enumerate(file_list):
    if os.path.isfile(filename):  # Only process files
        try:
            # Compute the MD5 hash for the file
            filehash = file_hash(filename)
            
            # If the hash is not in the dictionary, add it (it’s an original)
            if filehash not in hash_keys:
                hash_keys[filehash] = index
            else:
                # If the hash exists, it is a duplicate, so add it to the duplicates list
                duplicates.append((index, hash_keys[filehash]))
        except Exception as e:
            print(f"Error processing {filename}: {e}")
```

```python
# Print the duplicates (for debugging)
print("Duplicate files found:")
print(duplicates)
print(f"Total number of duplicate files: {len(duplicates)}")
```

```python
# Move the original images to the new folder
for index in hash_keys.values():
    try:
        # Get the filename of the original image
        original_image = file_list[index]
        
        # Define the full path for the original image
        original_path = os.path.join(os.getcwd(), original_image)
        
        # Define the destination path
        destination_path = os.path.join(destination_folder, original_image)
        
        # Move the original image to the destination folder
        shutil.move(original_path, destination_path)
        print(f"Moved {original_image} to {destination_folder}")
    except Exception as e:
        print(f"Error moving {file_list[index]}: {e}")
```
