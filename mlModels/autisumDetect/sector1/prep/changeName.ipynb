```python
import os
import shutil

# Directories
input_directory = r'C:\Users\Yasindu\Desktop\Stuff\1.SLIIT\Research\Project\test_data\prepImg\step2\Non_Autistic'  # Change this to your input directory
output_directory = r'C:\Users\Yasindu\Desktop\Stuff\1.SLIIT\Research\Project\test_data\prepImg\New folder\Non_Autistic'  # Change this to your output directory

# Ensure the output directory exists
if not os.path.exists(output_directory):
    os.makedirs(output_directory)

# Get a list of all the files in the input directory
files = os.listdir(input_directory)

# Sort the files to make sure they are processed in order (optional)
files.sort()

# Counter for the new filenames
counter = 00

# Loop through each file
for file in files:
    file_path = os.path.join(input_directory, file)  # Full path of the file
    if os.path.isfile(file_path):  # Ensure it's a file
        # Generate the new file name
        new_file_name = f"{counter}.jpg"  # You can change the extension as needed
        new_file_path = os.path.join(output_directory, new_file_name)  # New file path

        # Rename and move the file
        shutil.move(file_path, new_file_path)
        
        print(f"Moved {file} to {new_file_name}")

        # Increment the counter for the next file
        counter += 1
```
