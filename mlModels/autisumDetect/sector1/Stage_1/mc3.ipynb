```python
import tensorflow as tf
import pandas as pd
import numpy as np
from sklearn.metrics import confusion_matrix, classification_report, f1_score, precision_score
```

```python
# Configuration
model_path = r"C:\Users\Yasindu\Desktop\Stuff\1.SLIIT\Research\Project\models\model_3.h5"
test_dir = r"C:\Users\Yasindu\Desktop\Stuff\1.SLIIT\Research\Project\ModelTestingIMGS\model3"

# 1. Load Model & Dataset
# Clear session to help prevent GPU memory errors from previous runs
tf.keras.backend.clear_session()
model = tf.keras.models.load_model(model_path)

test_ds = tf.keras.utils.image_dataset_from_directory(
    test_dir, image_size=(224, 224), batch_size=8, label_mode='binary', shuffle=False
)
test_ds = test_ds.map(lambda x, y: (x/255.0, 1-y))

# 2. Inference
y_true, y_pred = [], []
for images, labels in test_ds:
    # Unpacking the unified model structure (Classification, Features)
    preds, _ = model.predict(images, verbose=0)
    y_true.extend(labels.numpy().flatten())
    y_pred.extend((preds > 0.5).astype(int).flatten())

# 3. Calculate Advanced Metrics
tn, fp, fn, tp = confusion_matrix(y_true, y_pred).ravel()

accuracy = (tp + tn) / len(y_true)
precision = tp / (tp + fp) if (tp + fp) > 0 else 0
recall = tp / (tp + fn) if (tp + fn) > 0 else 0  # Sensitivity
specificity = tn / (tn + fp) if (tn + fp) > 0 else 0
f1 = 2 * (precision * recall) / (precision + recall) if (precision + recall) > 0 else 0

# 4. Display Results
print("\n" + "="*30)
print(f"DETAILED RESULTS: MODEL 3")
print("="*30)
print(f"True Positives (TP):  {tp}")
print(f"True Negatives (TN):  {tn}")
print(f"False Positives (FP): {fp}")
print(f"False Negatives (FN): {fn}")
print("-" * 30)
print(f"Accuracy:    {accuracy*100:.2f}%")
print(f"Precision:   {precision*100:.2f}%")
print(f"Recall:      {recall*100:.2f}% (Sensitivity)")
print(f"Specificity: {specificity*100:.2f}%")
print(f"F1-Score:    {f1*100:.2f}%")
print("-" * 30)

print("\nClassification Report:")
print(classification_report(y_true, y_pred, target_names=['Non-Autistic', 'Autistic']))
```
